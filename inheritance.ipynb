## Inheritance in Class

#### Single Inheritance

```python
# Parent Class (Super Class)
class Person:
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def introduce(self):
        print(f'Hi there my name is {self.name} and I am {self.age} years old')
```

```python
# Child class - Inherit all the properties and methods from Person class
class Employee(Person):
    
    def __init__(self, name, age, company, occupation):
        super().__init__(name, age)
        self.company = company
        self.occupation = occupation
        
    def work(self):
        print(f'I work as a {self.occupation} in {self.company}.')
```

## Multilevel Inheritance

1. Parent Class
2. Child Class
3. Grandchild Class


```python
class Employee:
    
    def __init__(self, emp_id, name):
        self.emp_id = emp_id
        self.name = name
        
    def get_employee_info(self):
        print(f'Employee Id : {self.emp_id} , Name : {self.name}')
```

```python
class Manager(Employee):
    
    def __init__(self, emp_id, name, department):
        super().__init__(emp_id, name)
        self.department = department
        
    def get_manager_info(self):
        super().get_employee_info()
        print(f'Department : {self.department}')
```

```python
class ProjectManager(Manager):
    
    def __init__(self, emp_id, name, department, project):
        super().__init__(emp_id, name, department)
        self.project = project
        
    def get_project_manager_info(self):
        super().get_manager_info()
        print(f'Project : {self.project}')
```

## Hierarchical Inheritance

```python
class user :

    def __init__(self, username, email_id) :
        self.username = username
        self.email_id = email_id

    def get_user_info(self) :
        print (f'Username : {self.username}, email_id : {self.email_id}')
```

```python
class admin(user) :

    def __init__(self,username,email_id,access) :
        super().__init__(username, email_id)
        self.access = access

    def get_admin_access(self) :
        print(f'Access Rights : {self.access}')
```

```python
class RegularUser(user) :

    def __init__(self, username, email_id, subscription_type) :
        super().__init__(username,email_id)
        self.subscriptipn = subscription_type

    def get_reg_user_subscription(self) :
        print(f'Subscription type : {self.subscription_type}')
```
